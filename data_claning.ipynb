## Verification of Candidate Code Uniqueness
Following the verification of student codes, it was observed that they are not unique. This is because some students have taken the BAC exam multiple times.

```python
def non_unique_values(df, column_name):

  if df[column_name].is_unique:
    print('The IDs values are unique')
  else:
    print('The IDs values are NOT unique')

  print(df[column_name].value_counts())


non_unique_values(df, "Cod unic candidat")
```

## Splitting Data into Personal Details and Exam Results
Now we need to create two dataframes out of this one: the first will contain the personal details, and the second will contain the exam's results.

### The Personal Details Dataframe

```python
# Unitate (SIIIR) reprezunta codul liceului si face legatura cu adresa.csv
persoana = df_promovat[['Cod unic candidat', 'Sex', 'Specializare', 'Profil', 'Fileira', 'Unitate (SIIIR)', 'An', 'Sesiune']]
```

```python
persoana.head(5)
```

```python
persoana = persoana.rename(columns={
    'Cod unic candidat': 'cod_candidat',
    'Sex': 'sex',
    'Specializare': 'specializare',
    'Profil': 'profil',
    'Fileira': 'fileira',
    'Unitate (SIIIR)': 'siiir',
    'An': 'an',
    'Sesiune': 'sesiune'
})
```

```python
rezultate_bac_initial = df_promovat[['Cod unic candidat', 'Subiect ea', 'Subiect eb', 'Subiect ec', 'Subiect ed',
                    'NOTA_EA', 'NOTA_EB', 'NOTA_EC', 'NOTA_ED', 'CONTESTATIE_EA', 'NOTA_CONTESTATIE_EA',
                    'CONTESTATIE_EB', 'NOTA_CONTESTATIE_EB', 'CONTESTATIE_EC', 'NOTA_CONTESTATIE_EC',
                    'CONTESTATIE_ED', 'NOTA_CONTESTATIE_ED', 'STATUS', 'Medie', 'An', 'Sesiune']]
```

```python
rezultate_bac = rezultate_bac_initial.copy()
```

```python
rezultate_bac.columns
```

We create a function that analyzes if the candidate has contested the grades. If true, then the final grade of the candidate will be the contested grade.

```python
note_finale = ['NOTA_EA', 'NOTA_EB', 'NOTA_EC', 'NOTA_ED']
contestatie = ['CONTESTATIE_EA', 'CONTESTATIE_EB', 'CONTESTATIE_EC', 'CONTESTATIE_ED']
note_contestatie = ['NOTA_CONTESTATIE_EA', 'NOTA_CONTESTATIE_EB', 'NOTA_CONTESTATIE_EC', 'NOTA_CONTESTATIE_ED']
```

```python
for i in range(len(note_finale)):
  rezultate_bac[note_finale[i]] = rezultate_bac.apply(

      lambda x: x[note_contestatie[i]] if x[contestatie[i]] == 'Da' else x[note_finale[i]], axis=1

  )
```

```python
rezultate_bac.columns
```

```python
rezultate_bac2 = rezultate_bac[['Cod unic candidat', 'Subiect ea', 'Subiect eb', 'Subiect ec',
                               'Subiect ed', 'NOTA_EA', 'NOTA_EB', 'NOTA_EC', 'NOTA_ED',
                               'STATUS', 'Medie', 'An', 'Sesiune']]
```

```python
print(rezultate_bac2.head())
```

```python
rezultate_bac2 = rezultate_bac2.rename(columns = {
    'Cod unic candidat': 'cod_candidat',
    'Subiect ea': 'subiect_ea',
    'Subiect eb': 'subiect_eb',
    'Subiect ec': 'subiect_ec',
    'Subiect ed': 'subiect_ed',
    'NOTA_EA': 'nota_ea',
    'NOTA_EB': 'nota_eb',
    'NOTA_EC': 'nota_ec',
    'NOTA_ED': 'nota_ed',
    'STATUS': 'status',
    'Medie': 'medie',
    'An': 'an',
    'Sesiune': 'sesiune'
})
```

```python
rezultate_bac2.columns
```

We will also add a new column that contains a list of the subjects taken at the BAC for each student.

```python
rezultate_bac2['materii_bac'] = rezultate_bac2[['subiect_ea', 'subiect_eb', 'subiect_ec', 'subiect_ed']].apply(
    lambda row: str([materie for materie in row.dropna().astype(str)]), axis=1
)
```
